```python
# 1] HANDLING MISSING VALUES AND DUPLICATED DATASETS
# -> it has

#  1. Detecting missing values - df.isnull()   df.notnull()
#  2. Droping missing values - df.dropna()   df.dropna(axis = 1)
#  3. Filling missing values - df.fillna()
#  4. Replacing missing values - df.interpolate()

#  5. Detecting duplicated values - df.duplicated()
#  6. Droping duplicated values - df.drop_duplicates()
```

```python
import pandas as pd
import numpy as np
```

```python
data  = {          # improper data                                                 
    
    'Name': ['Atharv', 'Akashh', 'Talwinder', 'Atharv', None, 'Akashh', 'Rehman', ''],
    'Age': [21, 23, 22, 21, np.nan, 23, 26, np.nan],
    'City': ['Pune', 'Kolkata', 'Hydrabad', 'Pune', 'Mumbai', None, 'Agartala', ''],
    'Email': ['atharv@example.com', 'akashh@example.com', None, 'atharv@example.com', 'talwinder@example.com', None, 'rehman@example.com', '']
}

df = pd.DataFrame(data)
df
```

```python
# Now we handle missing values first
```

```python
df.isnull()
```

```python
df.sort_index()
```

```python
df.dropna(axis = 0)
```

```python
df
```

```python
df["Age"].fillna("-")
```

```python
df
```

```python
df["Email"].fillna("Unknown")
```

```python
df["Age"].interpolate() 
```

```python
df
```

```python
df.duplicated(subset = ["Name"])
```

```python
df.duplicated(subset = ["Email"])
```

```python
df.drop_duplicates(subset = ["Email"])
```

```python
df.drop_duplicates(keep = 'first')
```

```python
df.drop_duplicates(keep = 'last')
```

```python
# we use number of methods for filtering and selecting data those are
# .loc[]  .iloc[]   & | ~ .query()  .isin() .where()   .head()  .tail()  .sample()  .str()
```

```python
# how to select specific columns 
```

```python
datas = {

     "EmployeeID": [101, 102, 103, 104, 105],
    "Name": ["Atharv", "Akashh", "Talwinder", "Diljit", "Rehman"],
    "Department": ["HR", "IT", "Finance", "IT", "HR"],
    "Salary": [50000, 70000, 60000, 72000, 58000],
    "JoiningDate": ["2022-01-15", "2021-06-12", "2020-08-20", "2023-03-05", "2019-11-10"],
    "FullTime": [True, True, False, True, False]
    

    
}

df = pd.DataFrame(datas)
df
```

```python
df[["Name"]]
```

```python
df[["EmployeeID", "Name"]]
```

```python
# now how to select specific rows
```

```python
df[df["Salary"]>50000]         # we basically put conditions to show a whole row
```

```python
df[df["Name"] == "Diljit" ]
```

```python
# Filtering with Mulitple Connditions
```

```python
df[df["Department"] == "IT"]
```

```python
df[(df["Department"] == "IT") & (df["Name"] == "Diljit")]
```

```python
df[(df["Salary"]> 50000) | (df["Department"] == "IT")]
```

```python
 # using .query()           -> more comfortable filtering
```

```python
df.query("Name == 'Akashh'")
```

```python
df.query("Salary > 60000 & JoiningDate > '2020-1-12'")
```

```python
# using .where()
```

```python
df.where(df["Salary"]> 50000)              # it will show NaN to those which are filtered out 
```

```python
df.where(df["JoiningDate"] > '2020-01-12')
```

```python
df.where((df["JoiningDate"] > '2020-01-12') & (df["Department"] == "IT") )
```
